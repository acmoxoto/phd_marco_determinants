```python
from pandas_ods_reader import read_ods
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

base_path = "Dados_19_11_ML.ods"

sheet_index2 = 2  # log data
sheet_index1 = 1  # linear data

data_log = read_ods(base_path , sheet_index2)   # log data
data_lin = read_ods(base_path , sheet_index1)   # linear data
data_lin['ID'] = data_lin['ID'].astype('int')
data_log['ID'] = data_log['ID'].astype('int')
```

```python
pais = data_log.loc[:,"País"]
ids = data_log.loc[:,"ID"]  #.astype('int')

dict_country_cat = {}
for A, B in zip(pais,ids):
    dict_country_cat[A] = B
    
```

```python
data_lin.pop('País')
#data_lin.pop('ID')
data_lincountries = data_log.pop('País')
#keys = data_log.pop('ID')
data_lin
```

```python
#X3 = data_log[data_log.columns].iloc[: , :].values
X3 = data_log[data_log.columns].iloc[: , :].values

inertia = []
for n in range(1, 15):
    algorithm = (KMeans(n_clusters = n, init='k-means++', n_init = 10, max_iter=500, 
                        tol=0.0001, random_state=111, algorithm='elkan'))
    algorithm.fit(X3)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (15 , 6))
plt.plot(np.arange(1, 15), inertia, 'o')
plt.plot(np.arange(1, 15), inertia, '-', alpha = 0.5)
plt.xlabel('Number of clusters'), plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X3)
principalDf = pd.DataFrame(data = principalComponents, columns = ['Component 1', 'Component 2']) #, 'Component 3', 'Component 4'])
```

```python
fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('x-axis', fontsize = 15)
ax.set_ylabel('y-axis', fontsize = 15)
ax.set_title('2D component PCA', fontsize = 20)
ax.scatter(principalDf['Component 1'], principalDf['Component 2'])
plt.xticks([])
plt.yticks([])
```

```python
clusters={}
percen=0
for i in range(3):
    percen+=33.3
    clusters[i] = np.percentile(principalDf, percen, axis=0)
```

```python
fig = plt.figure(figsize = (6,4)) 
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15) 
ax.set_ylabel('Principal Component 2', fontsize = 15) 
ax.set_title('2D component PCA', fontsize = 20) 
ax.scatter(principalDf['Component 1'],principalDf['Component 2']) 
xposition = [clusters[0][0],clusters[1][0],clusters[2][0]] 
for xc in xposition: 
    plt.axvline(x=xc, color='k', linestyle='--')
plt.show()
```

```python
###################
#list for holding the data for each part

cluster_value1=[] 
cluster_value2=[]
cluster_value3=[]

for p in principalDf['Component 1']:
    if p<=clusters[0][0]:
        cluster_value1.append(p)
    elif (p>clusters[0][0] and p<=clusters[1][0]):
        cluster_value2.append(p)
    elif (p>clusters[1][0] and p<=clusters[2][0]):
        cluster_value3.append(p) 
        
#############################
data1=principalDf[principalDf['Component 1'].isin(cluster_value1)]
data2=principalDf[principalDf['Component 1'].isin(cluster_value2)]
data3=principalDf[principalDf['Component 1'].isin(cluster_value3)]

###############################
#merging the segregated part to the main data for extracting the real data.
    
first=pd.merge(data_log, data1, left_index=True, right_index=True).iloc[:,:38]
second=pd.merge(data_log, data2,left_index=True, right_index=True).iloc[:,:38]
third=pd.merge(data_log, data3, left_index=True, right_index=True).iloc[:,:38]

    # ################################
```

```python
iteration=[]
time_execution=[]

import time
for i in range(10):
    start=time.time()
    centers=[]
    j=0
    for i in [first,second,third]:
        #calculating mean values of each segment for finding the optimum cluster centroids 
        centers.append(np.mean(i).values) 
    # ##############################

    centers=np.array(centers)

    #feeding the centroids to the main k-means clustering algorithm
    kmeans = KMeans(3, init = centers,max_iter = 100000)
    kmeans.fit(data_log)

    end=time.time()

    execution_time=end-start
    no_of_iteration=kmeans.n_iter_

    iteration.append(no_of_iteration)
    time_execution.append(execution_time)

# print((execution_time))
# print(no_of_iteration)
```

```python
# Multidimensional Cluster Visualization with Parallel Coordinates

predict = kmeans.predict(data_log)
data_log['cluster'] = predict
pd.plotting.parallel_coordinates(data_log, 'cluster')
```
