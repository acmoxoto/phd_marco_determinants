```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo (substitua pelo seu DataFrame real)
data = {
    'Category': ['Commerce & Retail', 'Education & Research', 'Energy & Sustainability'],
    'Raised': ['1.237.967.447,00', '226.058.616,00', '1.444.074.656,00'],
    'ESG': [0.102, 0.052, 0.536],
    'E': [0, 0.002, 0.278],
    'S': [0.05, 0.04, 0.247],
    'G': [102, 30, 111]
}

df = pd.DataFrame(data)

# Converter para numéricos
def convert_currency(val):
    return float(val.replace('.', '').replace(',', '.'))

df['Raised'] = df['Raised'].apply(convert_currency)
df['category_code'] = df['Category'].astype('category').cat.codes

# Criar o gráfico
plt.figure(figsize=(14, 10))
scatter = plt.scatter(
    x=df['Raised'],
    y=df['category_code'],
    s=df['ESG'] * 2000,  # Aumentei o multiplicador para tamanhos visíveis
    c=df['E'],  # Usando a coluna E para cores
    alpha=0.7,
    cmap='plasma',
    edgecolors='black'
)

# Personalização avançada
plt.title('Bubble Chart: Investimento por Categoria', fontsize=16, pad=20)
plt.xlabel('Valor Arrecadado (USD)', fontsize=12)
plt.ylabel('Categoria', fontsize=12)

# Eixo Y com nomes
plt.yticks(
    ticks=df['category_code'],
    labels=df['Category'],
    rotation=45,
    ha='right',
    fontsize=10
)

# Adicionar elementos informativos
cbar = plt.colorbar(scatter)
cbar.set_label('Índice Ambiental (E)', fontsize=10)

# Anotar valores do G
for i, row in df.iterrows():
    plt.annotate(f"G: {row['G']}", 
                 (row['Raised'], row['category_code']),
                 textcoords="offset points",
                 xytext=(15,0),
                 ha='left')

# Grade e escala
plt.grid(True, alpha=0.3, linestyle='--')
plt.xscale('log')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Legenda de tamanho
sizes = [0.1, 0.3, 0.5]
plt.legend(
    handles=[plt.scatter([], [], s=s*2000, edgecolor='black', alpha=0.7) for s in sizes],
    labels=[f'ESG: {s}' for s in sizes],
    title='Tamanho das Bolhas',
    loc='upper right',
    bbox_to_anchor=(1.35, 1)
)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Dados de exemplo
data = {
    'Category': ['Commerce & Retail', 'Education & Research', 'Energy & Sustainability',
                 'Entertainment & Media', 'Finance & Economics', 'Governance & Legal',
                 'Logistics & Transportation', 'Health & Wellbeing', 'Social & Community',
                 'Technology & Innovation'],
    'Raised': ['1.237.967.447,00', '226.058.616,00', '1.444.074.656,00',
               '3.908.456.243,00', '10.298.473.446,35', '595.565.422,00',
               '985.622.978,00', '818.943.612,00', '953.694.553,00',
               '5.899.891.106,00'],
    'ESG': [0.102, 0.052, 0.536, 0.225, 0.878, 0.062, 0.127, 0.181, 0.093, 0.373],
    'E': [0, 0.002, 0.278, 0.024, 0.048, 0.002, 0.01, 0.012, 0.002, 0.02],
    'S': [0.05, 0.04, 0.247, 0.133, 0.467, 0.025, 0.063, 0.118, 0.061, 0.203],
    'G': [102, 30, 111, 330, 826, 70, 118, 89, 103, 382]
}

df = pd.DataFrame(data)

# Função de conversão de valores monetários
def convert_currency(val):
    try:
        return float(val.replace('.', '').replace(',', '.'))
    except:
        return 0.0

# Converter colunas numéricas
df['Raised'] = df['Raised'].apply(convert_currency)
df['category_code'] = df['Category'].astype('category').cat.codes

# Configurações estéticas
mpl.rcParams['font.family'] = 'Segoe UI'
plt.style.use('ggplot')

# Criar figura e eixo
fig, ax = plt.subplots(figsize=(18, 10), dpi=100)

# Criar o scatter plot
scatter = ax.scatter(
    x=df['Raised'],
    y=df['category_code'],
    s=df['ESG'] * 2500,  # Tamanho baseado no ESG
    c=df['G'],           # Cor baseada no Governance (G)
    cmap='coolwarm',
    alpha=0.8,
    edgecolor='black',
    linewidth=0.5,
    zorder=3
)

# Personalização do eixo Y
ax.set_yticks(df['category_code'])
ax.set_yticklabels(df['Category'], fontsize=10, rotation=30, ha='right')
ax.yaxis.grid(True, linestyle='--', alpha=0.4)

# Personalização do eixo X
ax.set_xlabel('Capital Arrecadado (USD)', fontsize=12, labelpad=10)
ax.set_xscale('log')
ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, _: f'${x/1e9:.1f}B'))
ax.xaxis.grid(True, linestyle='--', alpha=0.4)

# Barra de cores para Governance
cbar = plt.colorbar(scatter, ax=ax, pad=0.01)
cbar.set_label('Governança (G)', fontsize=10)
cbar.ax.tick_params(labelsize=8)

# Legenda de tamanho para ESG
kw = dict(prop="sizes", num=5, color="gray", 
          func=lambda s: f"ESG: {s/2500:.3f}", 
          fmt="{x:.0f}")
legend2 = ax.legend(*scatter.legend_elements(**kw),
                    title="Tamanho das Bolhas", 
                    loc="upper left",
                    bbox_to_anchor=(1, 1),
                    frameon=True,
                    fontsize=8)
ax.add_artist(legend2)

# Anotações para valores E e S
for i, row in df.iterrows():
    ax.annotate(f"E:{row['E']}\nS:{row['S']}", 
                (row['Raised'], row['category_code']),
                textcoords="offset points",
                xytext=(15,5),
                ha='left',
                va='bottom',
                fontsize=7,
                bbox=dict(boxstyle="round,pad=0.3", 
                          fc="white", 
                          ec="gray", 
                          lw=0.5))

# Elementos adicionais
plt.title('Análise de Investimentos por Categoria\n'
          'Tamanho: ESG Score | Cor: Governance Score | Anotações: E (Ambiental) e S (Social)',
          fontsize=14, pad=20)

# Ajustar layout
plt.tight_layout()
plt.subplots_adjust(right=0.75)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Dados de exemplo
data = {
    'Category': ['Commerce & Retail', 'Education & Research', 'Energy & Sustainability',
                 'Entertainment & Media', 'Finance & Economics', 'Governance & Legal',
                 'Logistics & Transportation', 'Health & Wellbeing', 'Social & Community',
                 'Technology & Innovation'],
    'Raised': ['1.237.967.447,00', '226.058.616,00', '1.444.074.656,00',
               '3.908.456.243,00', '10.298.473.446,35', '595.565.422,00',
               '985.622.978,00', '818.943.612,00', '953.694.553,00',
               '5.899.891.106,00'],
    'ESG': [0.102, 0.052, 0.536, 0.225, 0.878, 0.062, 0.127, 0.181, 0.093, 0.373],
    'E': [0, 0.002, 0.278, 0.024, 0.048, 0.002, 0.01, 0.012, 0.002, 0.02],
    'S': [0.05, 0.04, 0.247, 0.133, 0.467, 0.025, 0.063, 0.118, 0.061, 0.203],
    'G': [102, 30, 111, 330, 826, 70, 118, 89, 103, 382]
}

df = pd.DataFrame(data)

# ----------------------------------------------
# CORREÇÃO CRÍTICA: Função de conversão monetária
# ----------------------------------------------
def convert_currency(val):
    if isinstance(val, str):
        # Remove pontos (separadores de milhar) e substitui vírgula decimal por ponto
        cleaned = val.replace('.', '').replace(',', '.')
        return float(cleaned)
    return float(val)  # Se já for numérico, retorna direto

# Converter colunas numéricas
df['Raised'] = df['Raised'].apply(convert_currency)
df['Category'] = pd.Categorical(df['Category'], categories=df['Category'].unique())
df['category_code'] = df['Category'].cat.codes  # Códigos numéricos para as categorias

# ----------------------------------------------
# Verificação de dados (IMPORTANTE!)
# ----------------------------------------------
print("\nVerificação de Dados:")
print(df[['Raised', 'ESG', 'E', 'S', 'G']].describe())
print("\nAmostra dos Dados Convertidos:")
print(df.head(3))

# ----------------------------------------------
# Configuração do Gráfico
# ----------------------------------------------
mpl.rcParams['font.family'] = 'Segoe UI'
plt.style.use('seaborn-darkgrid')

fig, ax = plt.subplots(figsize=(18, 10), dpi=100)

# Criar o gráfico de bolhas
scatter = ax.scatter(
    x=df['Raised'],
    y=df['category_code'],
    s=df['ESG'] * 2500,
    c=df['G'],
    cmap='viridis',
    alpha=0.8,
    edgecolor='black',
    linewidth=0.5,
    zorder=3
)

# ----------------------------------------------
# Personalização dos Eixos
# ----------------------------------------------
ax.set_yticks(df['category_code'])
ax.set_yticklabels(df['Category'], fontsize=10, rotation=30, ha='right')
ax.set_xlabel('Capital Arrecadado (USD)', fontsize=12, fontweight='bold')
ax.set_ylabel('Categoria', fontsize=12, fontweight='bold')
ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, _: f'${x/1e9:.1f}B'))
ax.set_xscale('log')  # Escala logarítmica para melhor distribuição

# ----------------------------------------------
# Elementos Adicionais
# ----------------------------------------------
# Barra de cores para Governance (G)
cbar = plt.colorbar(scatter, ax=ax, pad=0.01)
cbar.set_label('Governança (G)', fontsize=10, fontweight='bold')

# Legenda de tamanho para ESG
legend_labels = [f"ESG: {esg}" for esg in df['ESG'].unique()]
legend_sizes = [esg * 2500 for esg in df['ESG'].unique()]
legend_handles = [plt.scatter([], [], s=size, edgecolor='black', alpha=0.8) for size in legend_sizes]
ax.legend(legend_handles, legend_labels, title='Tamanho das Bolhas', loc='upper left', bbox_to_anchor=(1, 1))

# Anotações
for i, row in df.iterrows():
    ax.annotate(
        f"E: {row['E']}\nS: {row['S']}",
        xy=(row['Raised'], row['category_code']),
        xytext=(20, 0),
        textcoords='offset points',
        ha='left',
        va='center',
        fontsize=8,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", lw=0.5)
    )

# Título
plt.title('Análise de Investimentos por Categoria', fontsize=16, fontweight='bold', pad=20)

# Ajustes Finais
plt.tight_layout()
plt.subplots_adjust(right=0.75)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criar dados de exemplo (substitua com seus dados reais)
data = pd.DataFrame({
    'Raised': np.random.randint(1, 100, 30),
    'Category': np.random.choice(['E', 'S', 'G'], 30),
    'Investimentos': np.random.randint(10, 500, 30),
    'Empresa': [f'Empresa {i}' for i in range(1, 31)]
})

# Configurações estéticas
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

# Criar gráfico de bolhas
scatter = sns.scatterplot(
    data=data,
    x='Raised',
    y='Category',
    size='Investimentos',
    sizes=(50, 500),  # Tamanho mínimo e máximo das bolhas
    hue='Category',   # Cores por categoria
    palette={'E': '#1f77b4', 'S': '#ff7f0e', 'G': '#2ca02c'},  # Cores customizadas
    alpha=0.7,       # Transparência
    edgecolor='black',  # Borda das bolhas
    linewidth=0.5     # Espessura da borda
)

# Adicionar rótulos para as maiores bolhas
for line in range(data.shape[0]):
    if data['Investimentos'].iloc[line] > 300:  # Só rotula investimentos maiores
        scatter.text(data['Raised'].iloc[line]+1, 
                    data['Category'].iloc[line], 
                    data['Empresa'].iloc[line],
                    horizontalalignment='left', 
                    size='medium',
                    color='black')

# Ajustes finais
plt.title('Gráfico de Bolhas: Captação por Categoria', fontsize=16, pad=20)
plt.xlabel('Valor Captado (em milhões)', fontsize=12)
plt.ylabel('Categoria ESG', fontsize=12)
plt.legend(title='Investimentos', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionar linha de média
mean_raised = data['Raised'].mean()
plt.axvline(mean_raised, color='gray', linestyle='--', alpha=0.5)
plt.text(mean_raised+2, 2.5, f'Média: {mean_raised:.1f}', color='gray')

sns.despine(left=True)  # Remove borda esquerda
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

## 1. Criar dados de exemplo (substitua com seus dados reais)
data = pd.DataFrame({
    'Raised': np.random.uniform(1, 100, 50),  # Valores contínuos para X
    'Category': np.random.choice(['E', 'S', 'G'], 50),  # Categorias para Y
    'Size': np.random.uniform(100, 1000, 50),  # Tamanho das bolhas
    'Performance': np.random.uniform(0, 1, 50)  # Cor das bolhas (opcional)
})

## 2. Configuração do estilo
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

## 3. Criar gráfico de bolhas
bubble = sns.scatterplot(
    data=data,
    x='Raised',
    y='Category',
    size='Size',
    hue='Performance',  # Opcional: cor baseada em outra variável
    palette="viridis",  # Escala de cores
    sizes=(50, 500),    # Range de tamanhos
    alpha=0.8,          # Transparência
    edgecolor='black',  # Borda das bolhas
    linewidth=0.5       # Espessura da borda
)

## 4. Personalizações avançadas
# Título e rótulos
plt.title('Relação entre Captação (Raised) e Categoria ESG', fontsize=16, pad=20)
plt.xlabel('Valor Captado (US$ milhões)', fontsize=12)
plt.ylabel('Categoria', fontsize=12)

# Ajustar legenda
plt.legend(
    title='Tamanho/Performance',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    frameon=True
)

# Linha de média
mean_val = data['Raised'].mean()
plt.axvline(mean_val, color='red', linestyle='--', alpha=0.7)
plt.text(mean_val+2, 2.7, f'Média: {mean_val:.2f}M', color='red')

# Adicionar valor médio por categoria
for category in ['E', 'S', 'G']:
    cat_mean = data[data['Category'] == category]['Raised'].mean()
    plt.scatter(cat_mean, category, s=300, facecolors='none', edgecolors='black')
    plt.text(cat_mean, category, f'{cat_mean:.1f}', 
             ha='center', va='center', fontsize=10)

## 5. Ajustes finais
sns.despine(left=True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Prepare the data
data = pd.DataFrame({
    'Category': ['Commerce & Retail', 'Education & Research', 'Energy & Sustainability', 
                'Entertainment & Media', 'Finance & Economics', 'Governance & Legal',
                'Logistics & Transportation', 'Health & Wellbeing', 'Social & Community',
                'Technology & Innovation'],
    'Raised': [1237967447, 226058616, 1444074656, 3908456243, 10298473446.35, 
               595565422, 985622978, 818943612, 953694553, 5899891106],
    'ESG': [0.102, 0.052, 0.536, 0.225, 0.878, 0.062, 0.127, 0.181, 0.093, 0.373],
    'E': [0, 0.002, 0.278, 0.024, 0.048, 0.002, 0.01, 0.012, 0.002, 0.02],
    'S': [0.05, 0.04, 0.247, 0.133, 0.467, 0.025, 0.063, 0.118, 0.061, 0.203],
    'G': [102, 30, 111, 330, 826, 70, 118, 89, 103, 382]
})

# Convert raised amounts to billions for better readability
data['Raised (B)'] = data['Raised'] / 1e9

# Create a composite color metric (E+S normalized)
data['Color_Metric'] = (data['E'] + data['S']) / data[['E', 'S', 'G']].sum(axis=1)

# Set up the figure
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Create bubble plot
scatter = sns.scatterplot(
    data=data,
    x='Raised (B)',
    y='Category',
    size='ESG',
    hue='Color_Metric',
    sizes=(100, 1000),  # Bubble size range
    palette='viridis',   # Color palette
    alpha=0.8,
    edgecolor='black',
    linewidth=0.5
)

# Add value labels to bubbles
for i, row in data.iterrows():
    plt.text(row['Raised (B)'], 
             row['Category'], 
             f"${row['Raised (B)']:.1f}B\nESG:{row['ESG']:.2f}",
             ha='center', 
             va='center',
             fontsize=8,
             color='white' if row['Color_Metric'] > 0.3 else 'black')

# Customize appearance
plt.title('ESG Investments by Sector (Size = ESG Score, Color = E+S Composition)', 
          fontsize=16, pad=20)
plt.xlabel('Amount Raised (Billions USD)', fontsize=12)
plt.ylabel('Industry Sector', fontsize=12)
plt.xscale('log')  # Log scale for better visualization of wide range

# Improve legend
handles, labels = scatter.get_legend_handles_labels()
plt.legend(handles[1:], labels[1:], 
           title='ESG Metrics',
           bbox_to_anchor=(1.05, 1),
           loc='upper left')

# Add reference lines
mean_raised = data['Raised (B)'].mean()
plt.axvline(mean_raised, color='red', linestyle='--', alpha=0.5)
plt.text(mean_raised*1.1, 8, f'Mean: ${mean_raised:.2f}B', color='red')

plt.tight_layout()
plt.show()
```
